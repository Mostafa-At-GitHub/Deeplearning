# Skew

Skewness is a measure of the asymmetry of the probability distribution of a real valued rando variable about its mean.
The skewness can be positive or negative or undefined. 
The negative skew: the left tail is longer; the mass of the distribution is concentrated on the right of the figure. 
The distribution can be described as left-skewed, left-tailed or skewed to the left
The positive skew: The right tail is longer; the mass of the distribution is concentrated on the left of the figure. 
The median is the value separating the higher half of a data sample 

skew = (mean - median) / std

# Skew normal distribution
The further detail math is here

https://en.wikipedia.org/wiki/Skew_normal_distribution



```python
import numpy as np
from scipy.special import erf
from scipy import pi, sqrt, exp
import matplotlib.pyplot as plt 
%matplotlib inline
# standard normal probability density function
def pdf(x):
    return 1/sqrt(2*pi) * exp((-x**2) / 2)
# with the cumulative distribution function given by
def cdf(x):
    return (1 + erf(x / sqrt(2))) / 2
def skew(x,e=0,w=1,a=0):
# e is the location parameter, and the w is scale parameter
# and transform the variable x to t 
# a is the skewness variable, if the a > 0 then it's left skewed else if a < 0, then it's right skewed
    t = (x-e) / w
    return 2 / w * pdf(t) * cdf(a*t)
    # You can of course use the scipy.stats.norm versions
    # return 2 * norm.pdf(t) * norm.cdf(a*t)
x = np.linspace(-10, 10, 1000)
e = 1.0 # location
w = 2.0 # scale
Y = np.empty([9, len(x)])
Y.shape
for i in range(-4, 5):
    Y[i][:] = skew(x,e,w,i)
for i in range(-4, 5):
    plt.plot(x, Y[i])
```

# Kurtosis
"In probability thory and statics, kurtosis is measure of the "tailedness" of the probability distibution of a real valued random variable. In a similar way, skewness, kurtosis is a desciptor of the shape of a probability distribution and just as for skewness, there are different ways of quantifying it for a theoretical distribution and corresponding ways of estimation it fomr a sample from a population. "

The kurtosis is the fourth standardized moment, 

$$Kurt[X] = \frac{E[(X - \mu)^4]}{(E[(X - \mu)^2])^2} $$

Kurtosis is certainly not the location of where the peak is. It's already called the mode.

The sample kurtosis is correspondingly related to the mean fourth power of a standardized variable (in some cases it is scaled by a factor that goes to 1 in large samples).

that's sometimes called excess kurtosis (it's also the fourth cumulant). You should beware - when seeing the word 'kurtosis' you need to keep in mind this possibility that different people use the same word to refer to two different (but closely related) quantities.

Kurtosis is usually either described as peakedness* (say, how sharply curved the peak is. which was presumably the intent of choosing the word "kurtosis") or heavy-tailedness (often what people are interested in using it to measure), but in actual fact the usual fourth standardized moment measure doesn't quite measure either of those things.

Indeed, the first volume of Kendall and Stuart give counterexamples that show that higher kurtosis is not necessarily associated with either higher peak (in a standardized variable) or fatter tails (in rather similar way that the third moment doesn't quite measure what many people think it does).

However in many situations there's some tendency to be associated with both, in that greater peakedness and heavy tailedness often tend to be seen when kurtosis is higher.

We should simply beware thinking it is necessarily the case.

Kurtosis and skewness are strongly related; interpretation of kurtosis is somewhat easier when the distribution is nearly symmetric.

A Gaussian distribution has a kurtosis of 0
A flatter distribution has negative kurtosis,
A distribution more peaked than a gaussian distribution has positive kurtosis
